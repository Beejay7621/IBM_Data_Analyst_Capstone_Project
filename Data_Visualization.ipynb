# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df['ConvertedComp'].plot(kind = 'hist', figsize=(10, 6))
plt.title('Salary Distribution')
plt.ylabel('Frequency of Annual Salary')
plt.xlabel('Annual Salary (USD)')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
df['Age'].plot(kind = 'box', figsize=(10, 6))
plt.title('Distribution of Age')
plt.ylabel('Age Distribution')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df.plot(kind = 'scatter', x = 'Age', y = 'WorkWeekHrs', figsize = (10, 6))
plt.title('Plot of Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
ax = df.plot(kind='scatter',
                    x = 'WorkWeekHrs',
                    y = 'CodeRevHrs',
                    figsize = (10, 6),
                    alpha = 0.5,  # transparency
                    s = 'Age'
                    )

ax.set_title('Bubble Plot of Work Week Hours and Code Review Hours In Relation To Age')
ax.set_ylabel('Code Review Hours')
ax.set_xlabel('Work Week Hours')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# To view the content of the table DatabaseDesireNextYear
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# To create a dataframe for the top 5 databases that respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, Count(*) AS COUNT
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY,conn)

df.set_index('DatabaseDesireNextYear', inplace=True)
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode_list = [0.1, 0, 0, 0, 0] # ratio for each database with which to offset each wedge.
df['COUNT'].plot(kind='pie',
        figsize = (15, 6),
        autopct = '%1.1f%%', # add in percentages
        startangle = 90,     # start angle 90°
        shadow = True,       # add shadow 
        labels = None,
        pctdistance = 1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
        colors = colors_list,
        explode = explode_list # 'explode Postgresql which has the highest percentage
                            )

plt.title('Composition of The Top 5 Databases To Learn Next Year', y = 1.12)
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels = df.index, loc='upper left')

plt.show()
```

# Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY,conn)

# Create a stacked chart for age group 30 to 35

work_week = df['WorkWeekHrs'].median()
code_rev = df['CodeRevHrs'].median()

fig, ax = plt.subplots()
ax.bar('Age 30 - 35', work_week, label='WorkWeekHrs')
ax.bar('Age 30 - 35', code_rev, bottom=work_week, label='CodeRevHrs')
ax.legend()
# ax.legend(loc = 'upper left')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(QUERY,conn)

line_data = df.groupby('Age').median()
                       
line_data.plot(kind='line')

plt.title('Line Chart of Annual Salary for All Ages From 45 to 60')
plt.ylabel('Annual Salary (USD)')
plt.xlabel('Age')
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)

df.set_index('MainBranch', inplace=True)
df.plot(kind = 'barh', color = 'steelblue')
plt.xlabel('Frequency')
plt.ylabel('Main Branch')
plt.title('Freuqncy of Main Branch')
plt.show()
```
